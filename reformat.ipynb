```python
import pandas as pd
```

```python
df = pd.read_csv('profiles.csv')
df = df.replace('\n', ' ', regex=True)
df = df.rename(columns={"name": "Name", "email":"Email"}, errors="raise")
df = df.dropna(subset=['Email'])
df['Name'] = df['Name'].str.replace('"', "'")
df.to_csv('profiles_cleaned.csv', index=False)
display(df.info())
display(df)
```

```python
#split csv into chunks of 970 rows to be uploaded into sendy
chunk_size = 5000

# read the CSV file into a pandas dataframe
df = pd.read_csv('profiles_cleaned.csv')

# split the dataframe into chunks of 970 rows
chunks = [df[i:i+chunk_size] for i in range(0, len(df), chunk_size)]

# save each chunk to a separate CSV file
for i, chunk in enumerate(chunks):
    chunk.to_csv(f'./georgia_contacts/Idaho_Bar_Association_chunk_{i}.csv', index=False)
```
